```python
import pandas as pd
import numpy as np

np.random.seed(42)

n_rows = 20000

data = {
    "MERCHANT ID": np.arange(100000, 100000 + n_rows),
    "CURRENT PLAN": np.random.choice(["BASIC", "SHOPIFY"], n_rows, p=[0.7, 0.3]),
    "MONTHLY REVENUE": np.random.normal(5000, 3000, n_rows).round(2),
    "REVENUE GROWTH QOQ": np.random.normal(0.08, 0.15, n_rows).round(3),
    "NUMBER OF ORDERS": np.random.poisson(120, n_rows),
    "AVERAGE ORDER VALUE": np.random.normal(65, 25, n_rows).round(2),
    "ACTIVE DAYS LAST 30": np.random.randint(5, 30, n_rows),
    "APPS INSTALLED": np.random.randint(0, 25, n_rows),
    "SUPPORT TICKETS LAST 90": np.random.poisson(3, n_rows),
    "STAFF COUNT": np.random.randint(1, 50, n_rows),
    "COUNTRY": np.random.choice(
        ["USA", "CANADA", "UK", "FRANCE", "GERMANY", "SPAIN"], n_rows
    ),
    "MARKETING SPEND": np.random.normal(1200, 800, n_rows).round(2),
    "WEBSITE TRAFFIC": np.random.randint(500, 50000, n_rows),
    "CUSTOMER COUNT": np.random.randint(50, 2000, n_rows),
    "PAYMENT FAIL RATE": np.random.uniform(0, 0.15, n_rows).round(3),
}

df = pd.DataFrame(data)

# Create target variable (Upgrade logic)
df["UPGRADED NEXT QUARTER"] = (
    (df["MONTHLY REVENUE"] > 7000) &
    (df["REVENUE GROWTH QOQ"] > 0.1) &
    (df["APPS INSTALLED"] > 8)
).astype(int)

# Introduce missing values
for col in ["MARKETING SPEND", "STAFF COUNT", "REVENUE GROWTH QOQ"]:
    df.loc[df.sample(frac=0.08).index, col] = np.nan

# Introduce duplicates
df = pd.concat([df, df.sample(500)], ignore_index=True)

# Shuffle dataset
df = df.sample(frac=1).reset_index(drop=True)

# Save CSV
file_path = "/content/shopify_upgrade_dataset_raw.csv"
df.to_csv(file_path, index=False)

file_path
```

# Logistic Regression

```python
# Separate features (X) and target (y)
X = df.drop(columns=["upgraded_next_quarter"])
Y = df["upgraded_next_quarter"]
```

We separated the dataset into features (X) and the target variable (Y).
The goal is to predict whether a merchant will upgrade next quarter based on business activity metrics.

```python
# Split the data into training and testing sets
# Stratify ensures the target distribution (0/1) stays similar in both sets

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42,
    stratify=Y
)

X_train.shape, X_test.shape
```

We split the data into train/test sets to evaluate model performance on unseen data.
Stratified split was used to preserve the class distribution.

```python
# Build a preprocessing pipeline:
# - OneHotEncode categorical columns
# - Scale numerical columns
# This avoids data leakage and keeps transformations consistent

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline

# Identify categorical and numerical columns
cat_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()
num_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()

preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
    ]
)

cat_cols, num_cols
```

```python
# Train Logistic Regression as a baseline classification model
# class_weight="balanced" helps if the dataset is imbalanced (more 0s than 1s)

from sklearn.linear_model import LogisticRegression

model = Pipeline(steps=[
    ("preprocess", preprocess),
    ("clf", LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42))
])

model.fit(X_train, Y_train)
```

```python
# Predict class labels (0/1) and probabilities (propensity to upgrade)

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]  # probability of class 1 (upgrade)

y_pred[:10], _proba[:10]
```

```python
# Evaluate performance using standard classification metrics

from sklearn.metrics import classification_report, accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

Accuracy alone may be misleading when the classes are imbalanced.
Precision/Recall/F1 provide a better view of how well the model identifies likely upgraders.

```python
# Confusion Matrix helps interpret errors:
# - False Positives: predicted upgrade but actually no
# - False Negatives: predicted no upgrade but actually yes (very important for sales!)

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(Y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)

disp.plot()
plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

```python
# ROC-AUC measures the models ability to rank positives higher than negatives
from sklearn.metrics import roc_auc_score, RocCurveDisplay

auc = roc_auc_score(Y_test, y_proba)
print("ROC-AUC:", auc)

RocCurveDisplay.from_predictions(Y_test, y_proba)
plt.title("ROC Curve - Logistic Regression")
plt.show()
```

```python
# Adjust decision threshold to improve recall (catch more potential upgraders)
# This is common for sales/prioritization tasks

import numpy as np
from sklearn.metrics import accuracy_score, classification_report

threshold = 0.35
y_pred_custom = (y_proba >= threshold).astype(int)

print("Custom Threshold:", threshold)
print("Accuracy:", accuracy_score(Y_test, y_pred_custom))
print("\nClassification Report:\n")
print(classification_report(Y_test, y_pred_custom))
```

```python
# Extract feature names after preprocessing and show the most important coefficients
# Positive coefficient => increases probability of upgrading
# Negative coefficient => decreases probability of upgrading

import pandas as pd
import numpy as np

# Get feature names after preprocessing
ohe = model.named_steps["preprocess"].named_transformers_["cat"]
cat_feature_names = ohe.get_feature_names_out(cat_cols)

feature_names = np.concatenate([num_cols, cat_feature_names])

# Get coefficients
coefs = model.named_steps["clf"].coef_[0]

importance = pd.DataFrame({
    "feature": feature_names,
    "coef": coefs,
    "abs_coef": np.abs(coefs)
}).sort_values("abs_coef", ascending=False)

importance.head(15)
```

```python
# Build a prioritized list of merchants with the highest predicted upgrade probability

# We will generate probabilities for the test set and attach merchant_id (if available)
results = X_test.copy()
results["actual_upgrade"] = Y_test.values
results["propensity_to_upgrade"] = y_proba

# Sort by propensity score (highest first)
top_merchants = results.sort_values("propensity_to_upgrade", ascending=False).head(20)

top_merchants[["merchant_id", "propensity_to_upgrade", "actual_upgrade"]].head(20)
```

Logistic Regression requires numerical input features.

```python
# Scale numerical features to improve model convergence
# Logistic Regression performs better when features are on similar scales

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Feature scaling was applied using StandardScaler to ensure all numerical variables operate on comparable scales.

```python
# Extract model coefficients to identify key drivers
# Coefficients represent the impact of each feature on upgrade probability

import pandas as pd

feature_importance = pd.DataFrame({
    "feature": X_train.columns,
    "coefficient": log_reg.coef_[0]
})

# Sort by absolute impact
feature_importance["abs_coef"] = feature_importance["coefficient"].abs()
feature_importance = feature_importance.sort_values("abs_coef", ascending=False)

feature_importance.head(10)
```

```python
# Visualize the top 10 key drivers

import matplotlib.pyplot as plt

top_features = feature_importance.head(10)

plt.figure(figsize=(8,5))
plt.barh(top_features["feature"], top_features["coefficient"])
plt.axvline(0, color="black")
plt.title("Top 10 Key Drivers of Upgrade")
plt.xlabel("Logistic Regression Coefficient")
plt.ylabel("Feature")
plt.gca().invert_yaxis()
plt.show()
```

The key drivers analysis was performed using Logistic Regression coefficients.
Features with higher absolute coefficients have a stronger influence on upgrade decisions.
The results show that merchant growth metrics, engagement indicators, and platform usage are the strongest predictors of subscription upgrades.
These insights can help Shopifys sales team prioritize merchants with the highest upgrade potential.
